```python
from tqdm import tqdm
for f, x, y in tqdm(train_loader):
    for i in range(len(f)):
        dat = x[i]
        new_x = dat - np.mean(dat, axis=(1,2)).reshape(-1,1,1)
        new_x /= (np.std(new_x, axis=(1,2)).reshape(-1,1,1) + 1e-5)
        np.save(new_train_dir + f[i], new_x)
```
